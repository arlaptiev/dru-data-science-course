### Operation description

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5, 0.1, 0.8, 0.8]) # @params: left, right, width, height of canvas

axis.plot(x, y, 'r')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x,y, 'r')
```

### Draw another picture or graph inside another graph

```python
fig = plt.figure(figsize=(16,9), dpi=300) #dpi for quality

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # big axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # small canvas

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
fig, axes = plt.subplots()

axes.set_xlabel('x-label')
axes.set_ylabel('y-label')

axes.set_title('title')


axes.plot(x, x**2, 'y')
axes.plot(x, x**3, 'b')

axes.legend(['label1', 'label2'], loc=2)
```

### Line color, transparence, and more

```python
fig, axes = plt.subplots(dpi=150)

# aplha for transparency
axes.plot(x, x**2, color='red', alpha=0.5)
axes.plot(x, x+2, color='#1155dd', alpha=0.5)
axes.plot(x, x+3, color='green', alpha=0.1)
```

```python
fig, axes = plt.subplots(dpi=100)

axes.plot(x, x+1, color='blue', linewidth = 2)
axes.plot(x, x+2, color='blue', lw = 1)
axes.plot(x, x+3, color='blue', lw = 0.5)
axes.plot(x, x+4, color='blue', lw = 0.1)
```

```python
fig, axes = plt.subplots(dpi=100)

axes.plot(x, x+1, color='yellow', lw = 3, linestyle='-')
axes.plot(x, x+2, color='yellow', lw = 3, ls='-.')
axes.plot(x, x+3, color='yellow', lw = 3, ls=':')

line, = axes.plot(x, x+4, color='blue', lw=2)
line.set_dashes([5,10,15,10])

line, = axes.plot(x, x+5, color='blue', lw=2)
line.set_dashes([5,10,15,3])

line, = axes.plot(x, x+6, color='blue', lw=2)
line.set_dashes([5,10,4,10])
```

```python
fig, axes = plt.subplots(dpi=100)

axes.plot(x, x+1, color='blue')
axes.plot(x, x+2, color='blue', marker='o')
axes.plot(x, x+3, color='blue', marker='+')
axes.plot(x, x+4, color='blue', marker='s')
axes.plot(x, x+5, color='blue', marker='1')

axes.plot(x, x+6, color='blue', marker='o', markersize=13)
axes.plot(x, x+7, color='blue', marker='s', markerfacecolor='red')
```

### Limiting canvas grid and axis range

```python
fig, Axes = plt.subplots(1,2, figsize=(10,5))

Axes[0].plot(x, x**2, x, x**3, lw=2)
Axes[0].grid(True)

Axes[1].plot(x, x**2, x, x**3, lw=2)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
```

## Other 2D Graphics

```python
n = np.arange(6)

fig, axes = plt.subplots(1,4, figsize=(16,5))

axes[0].set_title('scatter')
axes[0].scatter(x, x + np.random.random(len(x)))

axes[1].set_title('step')
axes[1].step(n, n**2, lw=2)

axes[2].set_title('bar')
axes[2].bar(n, n**2, align='center', width=0.9, alpha=0.5)

axes[3].set_title('fill_between')
axes[3].fill_between(x, x**2, x**3, color='green', alpha=0.5)
```

### Radar chart

```python
fig = plt.figure(figsize=(6,6))

ax = fig.add_axes([0.0, 0.0, 1, 1], polar=True)
t = np.linspace(0, 2 * np.pi, 100)
ax.plot(t, t * 2, color='blue', lw=2)
```

### Histogram

```python
n = np.random.randn(100000)

fig, axes = plt.subplots(1,2, figsize=(12, 4))

axes[0].set_title("Default histogram")
axes[0].hist(n, bins=1000)

axes[1].set_title("Cumulative detailed histogram")
axes[1].hist(n, cumulative=True, bins=7, color='orange')
```

### Contour maps

```python
import matplotlib.cm as cm
```

```python
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1-Z2) * 2

print(X)
print(Y)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inlint=1, fontsize=10)
ax.set_title('countour map')
```
